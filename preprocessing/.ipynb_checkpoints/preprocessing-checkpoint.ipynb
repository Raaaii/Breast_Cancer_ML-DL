```python
scaler = StandardScaler()
```

```python
dataValues = scaler.fit_transform(data)
```

```python
dataValues
```

```python
data = KNNImputer(n_neighbors=4).fit_transform(dataValues)
```

```python
data
```

# Revesing the scaling because scaler should be fit only on the training data

```python
data = scaler.inverse_transform(data)
```

```python
data
```

```python
recurrenceAsTarget = LabelEncoder().fit_transform(recurrence.values)
```

```python
#mak new dataframe without time column
regDataNoTime = regData.drop(columns=["Time"])
regDataNoTime
```

```python
#choose best number of components for PCA from plot
from sklearn.decomposition import PCA
pca = PCA().fit(regDataNoTime)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.show()
```

```python
#We saw that 4 components explain 90% of the variance

pca = PCA(n_components=4)
pca.fit(regDataNoTime)
print("Explained variance: ", pca.explained_variance_ratio_)
print("Singular values: ", pca.singular_values_)
print("Components: ", pca.components_)
print("Mean: ", pca.mean_)
print("Number of components: ", pca.n_components_)
print("Number of samples: ", pca.n_samples_)
print("Number of features: ", pca.n_features_)
print("Noise variance: ", pca.noise_variance_)
```

```python
#make pca dataframe
pcaData = pd.DataFrame(pca.transform(regDataNoTime), columns=["PC1", "PC2", "PC3", "PC4"], index=index)
pcaData
```

```python
#make a heatmap of the components and their explained variance
plt.figure(figsize=(10,10))
sns.heatmap(pcaData.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()  
```

```python
#save pcdDf to csv
pcaData.to_csv("pcaData.csv")
pcaData
```

```python
#make a new dataframe with the target variable and the pca data
regDataPca = pd.concat([pcaData, regData["Time"]], axis=1)

#convert regDataPca to excel

regDataPca.to_excel("regDataPca.xlsx")
```
