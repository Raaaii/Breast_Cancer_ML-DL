# Decision Tree Regressor

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("regressionData.xlsx")
```

```python
df
```

## Prediction Target & Features

```python
y = df.Time
```

```python
features = df.columns.to_list()[:-1]
X = df[features]
```

## Building the Model

### 1. Split the data into training and validation data

```python
train_X, test_X, train_y, test_y = train_test_split(X, y)
```

### 2. Define a decision tree model with scikit-learn and fit it with the features and target variable

```python
# Define model
wisc_model = DecisionTreeRegressor()
```

```python
# Fit model
wisc_model.fit(train_X, train_y)
```

### 3. Assess model performance

```python
wisc_model.score(test_X, test_y)
```

```python
wisc_model.score(train_X, train_y)
```
